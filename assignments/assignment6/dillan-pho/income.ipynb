```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from ucimlrepo import fetch_ucirepo 
```

```python
#fetch dataset
census_income = fetch_ucirepo(id=20) 
```

```python

# data (as pandas dataframes) 
X = census_income.data.features 
Y = census_income.data.targets 
X = pd.concat([X, Y], axis=1)

# exploring dataset
#first 20 entries
X.head(20)
```

```python
#last 20 entries
X.tail(20)
```

```python
#correlation matrix
X['income'] = X['income'].map({'<=50K': 0, '>50K': 1})
numeric_df = X.select_dtypes(include=['float64', 'int64'])
corr_matrix = numeric_df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

## HYPOTHESIS 1:

- There are more people with bachelors degrees or higher make >50k, than those without.

```python
extracted_df = X.filter(['education', 'capital-gain', 'income'])

extracted_df['bachelors_and_above'] = extracted_df.apply(
    lambda row: 1 if row['education'] in ['Bachelors', 'Masters', 'Doctorate'] and row['income'] == 1 else 0, axis=1)

extracted_df['some_college_below'] = extracted_df.apply(
    lambda row: 1 if row['education'] in ['HS-grad', 'Some-college', 'Assoc-acdm', 'Assoc-voc', '11th', '10th', '9th',
                                          '7th-8th', '5th-6th', '1st-4th', 'Preschool'] and row['income'] == 1 else 0, axis=1)

print('Bachelors & Above (>50K):', extracted_df['bachelors_and_above'].sum())
print('Some College & Below (>50K):', extracted_df['some_college_below'].sum())
summary = extracted_df[['bachelors_and_above', 'some_college_below']].sum()
summary.plot(kind='bar')
plt.title('Counts of Education Levels with Income >50K')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

## Interpretations and Methodology

Based on the bar graph above, the evidence does not support my hypothesis of people with bachelors and higher degrees making more than 50k.

How I conducted my test was by focusing on two piece of the data set, education and income.

I created a new column called "bachelors_and_above" which focuses on if the person had a bachelors degree and had an income of over 50k,
then this would be marked 1 (for true). If it doesn't fufill the criteria then it gets marked 0 (false). I then summed the amount for this column to see the amount of people.

I applied the same methodology for those who didn't have bachelors degree.

What this means is that there were more people who had an education experience of some-college and below, that made more than 50k. 

This may imply that higher degrees do not mean more money, but more analysis is needed.

## HYPOTHESIS 2:

- Number of hours worked is a significant predictor of those who make over 50k.

```python
from scipy.stats import ttest_ind

group_over_50k = X[X['income'] == 1]['hours-per-week']
group_under_50k = X[X['income'] == 0]['hours-per-week']


t_stat, p_value = ttest_ind(group_over_50k, group_under_50k)

print('t-statistic:', t_stat)
print('p-value:', p_value)
```

## Interpretations and Methodology
To determine if the number of hours worked per week is a significant predictor of earning more than 50K annually, I conducted an independent t-test. This test compares the means of two independent groups: individuals earning >50K and those earning <=50K.

I utilized the 'income' column where those with >50k are mapped to 1 and those <=50k are mapped to 0. Then, I separated the dataset into two groups based on this mapping:

- group_over_50k: Individuals who earn more than 50K.
- group_under_50k: Individuals who earn 50K or less.

I extracted the 'hours-per-week' data for each group and performed an independent t-test using the ttest_ind function from the scipy.stats module. The t-test evaluates whether the mean number of hours worked per week differs significantly between the two income groups.

As a result we get a p-value of 0.0 and a t-statistic of 42.583873499437956.

The high t-statistic and a p-value of 0.0 indicate a significant difference in hours worked per week between the two income groups. This suggests that working more hours per week is strongly associated with earning more than $50K annually.
