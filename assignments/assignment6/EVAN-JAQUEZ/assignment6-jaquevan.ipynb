I will sort the dataset into two catagories using different csv files. The catagories will be red and white wine. With this I will compare key variables from the set using data visualization.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

red_wine = pd.read_csv('winequality-red.csv', sep=';')
white_wine = pd.read_csv('winequality-white.csv', sep=';')

print("Variables in the data set", white_wine.columns)

plt.figure(figsize=(10, 6))
plt.scatter(white_wine['alcohol'], white_wine['quality'], color='blue', label='White Wine', s=10)
plt.scatter(red_wine['alcohol'], red_wine['quality'], color='red', label='Red Wine', s=10)
plt.xlabel('Alcohol Content')
plt.ylabel('Quality')
plt.title("Alcohol Content vs. Quality in Red and White Wines")
plt.legend()


plt.figure(figsize=(10, 6))
plt.scatter(white_wine['sulphates'], white_wine['quality'], color='blue', label='White Wine', s=10)
plt.scatter(red_wine['sulphates'], red_wine['quality'], color='red', label='Red Wine', s=10)
plt.xlabel('Sulphates')
plt.ylabel('Quality')
plt.title("Sulphates vs. Quality in Red and White Wines")
plt.legend()

plt.figure(figsize=(10, 6))
plt.scatter(white_wine['pH'], white_wine['quality'], color='blue', label='White Wine', s=10)
plt.scatter(red_wine['pH'], red_wine['quality'], color='red', label='Red Wine', s=10)
plt.xlabel('pH Level')
plt.ylabel('Quality')
plt.title("pH vs. Quality in Red and White Wines")
plt.legend()
```

Above we can see all the variables that can be associated with red and white wine and we can forumlate hypothesises based on the data.

Figure 1.2

Red wines appear to have a wider range of sulphate values, extending up to around 2.0, compared to white wines, which tend to cluster around lower sulphate levels.

Figure 1.3
The scatter plot for pH vs. quality shows less clear correlation. However, most wines with high-quality ratings are within a pH range of about 3.0 to 3.4.

Hypothesis: 

Wines with higher levels of sulfure dioxide have lower quality ratings.

After doing some research on the variables in the data set I came across this recource : https://extension.okstate.edu/fact-sheets/understanding-free-sulfur-dioxide-fso2-in-wine.html

Sulfure Dioxide is used as a preservative in wine, so proving or disproving this hypothesis is interesting as we can see how preservatives impact the enjoyment of wine. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


red_wine = pd.read_csv('winequality-red.csv', sep=';')
white_wine = pd.read_csv('winequality-white.csv', sep=';')

print(white_wine.columns)


mean_total_so2_red = np.mean(red_wine['total sulfur dioxide'])
mean_total_so2_white = np.mean(white_wine['total sulfur dioxide'])

#  mean for each wine type to sort between high and low variables
high_so2_red = red_wine['total sulfur dioxide'] > mean_total_so2_red
low_so2_red = red_wine['total sulfur dioxide'] <= mean_total_so2_red
high_so2_white = white_wine['total sulfur dioxide'] > mean_total_so2_white
low_so2_white = white_wine['total sulfur dioxide'] <= mean_total_so2_white



# Red Wine low
plt.subplot(2, 2, 1)
plt.hist(red_wine[low_so2_red]['quality'], bins=6, color='pink', alpha=0.7)
plt.title('Red Wine - Low Sulfure Dioxide Quality')
plt.xlabel('Quality')
plt.ylabel('Frequency')

# Red Wine high
plt.subplot(2, 2, 2)
plt.hist(red_wine[high_so2_red]['quality'], bins=6, color='red', alpha=0.7)
plt.title('Red Wine - High Sulfure Dioxide Quality')
plt.xlabel('Quality')
plt.ylabel('Frequency')

# White Wine low
plt.subplot(2, 2, 3)
plt.hist(white_wine[low_so2_white]['quality'], bins=6, color='aqua', alpha=0.7)
plt.title('White Wine - Low Sulfure Dioxide Quality')
plt.xlabel('Quality')
plt.ylabel('Frequency')

# White Wine high
plt.subplot(2, 2, 4)
plt.hist(white_wine[high_so2_white]['quality'], bins=6, color='blue', alpha=0.7)
plt.title('White Wine - High Sulfure Dioxide Quality')
plt.xlabel('Quality')
plt.ylabel('Frequency')

plt.show()
```

For red win e the distribution of the quality scores for both low and  high sulfure dioxide was about the same, averaging between 5 and 6. 

In white wine, both high and low sulfur dioxide groups also have the majority of their quality scores around 6 and 7. 

However, in high levels of SO2, the data skews more to the left with more ratings below 6 than before. This conclusion can also be drawn from the red wine histogram where scores decreased on average. 

We can check this by checking the mean values:

```python
import pandas as pd
import numpy as np



high_so2_red = red_wine[red_wine['total sulfur dioxide'] > mean_total_so2_red]
low_so2_red = red_wine[red_wine['total sulfur dioxide'] <= mean_total_so2_red]
high_so2_white = white_wine[white_wine['total sulfur dioxide'] > mean_total_so2_white]
low_so2_white = white_wine[white_wine['total sulfur dioxide'] <= mean_total_so2_white]


print("Red Wine - Low Group: Mean Quality =", np.mean(low_so2_red['quality']))

print("Red Wine - High Group: Mean Quality =", np.mean(high_so2_red['quality']))

print("\nWhite Wine - Low Group: Mean Quality =", np.mean(low_so2_white['quality']))

print("White Wine - High Group: Mean Quality =", np.mean(high_so2_white['quality']))
```

By observing the histograms and the changes in the mean quality with high and low levels of Sufler Dioxide, this points in the direction that my original hypothesis was correct. While sulfure dioxide is needed for wine preservation these findings suggest that high levels of SO2 actually slightly decrease wine quality, negativly impacting the taste and aroma for the consumer.
