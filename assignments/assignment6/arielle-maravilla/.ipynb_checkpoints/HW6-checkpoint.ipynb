```python
# Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1.) Importing the dataset: Replace 'your_dataset_url' with the actual URL or file path
math_url = '/Users/ariellemaravilla/Desktop/DS219/spark-seprep/assignments/assignment6/arielle-maravilla/student+performance/student/student-mat.csv'
portugese_url = '/Users/ariellemaravilla/Desktop/DS219/spark-seprep/assignments/assignment6/arielle-maravilla/student+performance/student/student-por.csv'

math_df = pd.read_csv(math_url, delimiter=';')
portuguese_df = pd.read_csv(portugese_url, delimiter=';')



# 2.) Initial Exploratory Data Analysis (EDA): Explore the dataset's structure and summary statistics

#for math_df
math_df.head()      # Shows the first few rows
math_df.info()      # Shows column types and null values
math_df.describe()  # Shows summary statistics

#Repeat commands for portuguese_df
portuguese_df.head()
portuguese_df.info()
portuguese_df.describe()



# 3.) Data Visualization Exploration (Univariate, Bivaraite, Multivariate Analyses) 

#Univariate Analysis 1: Historgram

# Histogram (Portuguese): Distribution of final grades
plt.figure(figsize=(8, 5))
sns.histplot(portuguese_df['G3'], bins=10, kde=True)
plt.title('Distribution of Final Grades in Portuguese (G3)')
plt.xlabel('Final Grade (G3)')
plt.ylabel('Frequency')
plt.show()

# Histogram (Math): Distribution of final grades 
plt.figure(figsize=(8, 5))
sns.histplot(math_df['G3'], bins=10, kde=True)
plt.title('Distribution of Final Grades in Math (G3)')
plt.xlabel('Final Grade (G3)')
plt.ylabel('Frequency')
plt.show()



#Univariate Analysis 2: Pie Chart

# Pie Chart (Portuguese): Proportion of students with internet access
plt.figure(figsize=(6, 6))
internet_counts = portuguese_df['internet'].value_counts()
plt.pie(internet_counts, labels=internet_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Internet Access at Home (Portuguese)')
plt.show()

# Pie Chart (Math): Proportion of students with internet access
plt.figure(figsize=(6, 6))
internet_counts = math_df['internet'].value_counts()
plt.pie(internet_counts, labels=internet_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Internet Access at Home (Math)')
plt.show()



#Bivariate Analysis 1: Box Plot

# Box Plot (Portuguese): Final grades by gender 
plt.figure(figsize=(8, 5))
sns.boxplot(x='sex', y='G3', data=portuguese_df)
plt.title('Final Grades by Gender (Math)')
plt.xlabel('Gender')
plt.ylabel('Final Grade (G3)')
plt.show()

# Box Plot (Math): Final grades by gender
plt.figure(figsize=(8, 5))
sns.boxplot(x='sex', y='G3', data=math_df)
plt.title('Final Grades by Gender (Math)')
plt.xlabel('Gender')
plt.ylabel('Final Grade (G3)')
plt.show()



#Multivariate Analysis: Heatmap

# Correlation heatmap (Portuguese)
plt.figure(figsize=(10, 8))
correlation_vars = portuguese_df[['G1', 'G2', 'G3', 'studytime', 'absences', 'health']]
sns.heatmap(correlation_vars.corr(), annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Heatmap (Portuguese)')
plt.show()

# Correlation heatmap (Math)
plt.figure(figsize=(10, 8))
correlation_vars = math_df[['G1', 'G2', 'G3', 'studytime', 'absences', 'health']]
sns.heatmap(correlation_vars.corr(), annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Heatmap (Math)')
plt.show()


#More Heatmap Analysis:



# Select the 'Time Spent' variables and exam score variables
time_spent_with_grades = math_df[['traveltime', 'studytime', 'freetime', 'goout', 'absences', 'G1', 'G2', 'G3']]

# Calculate the correlation matrix
correlation_matrix = time_spent_with_grades.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap of Time Spent Variables with Exam Scores')
plt.show()





# Hypothesis Formulation
# Formulate at least two hypotheses based on what you observe in the visuals
```
