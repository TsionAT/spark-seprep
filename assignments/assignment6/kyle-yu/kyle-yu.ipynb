# Hypothesis

### Fixed Acidity and pH (white wine)
Null Hypothesis: There is no relationship between fixed acidity and pH in the white wine data.

Alternative Hypothesis: THere is a significant relationship between fixed acidity and pH in the white wine data.

```python
correlation1, p1 = stats.pearsonr(white["fixed acidity"], white["pH"])

print(f"Pearson Correlation Coefficient: {correlation1}")
print(f"P-value: {p1}")
```

We observed a moderately strong negative linear relationship between the fixed acidity and pH of the white wine dataset. Beacuse we observed a small p-value, we can conclude that it is extremely unlikely for this relationship to have occurred by raandom chance. Therefore we can reject the null hypothesis.

### Density and Alcohol (red wine)

Null hypothesis: There is no relationship between desnity and alcohol levels.

Alternative hypothesis: There is a significant relationship between desnity and alcohol levels.

```python
correlation2, p2 = stats.spearmanr(red["density"], red["alcohol"])

print(f"Spearman Correlation Coefficient: {correlation2}")
print(f"P-value: {p2}")
```

We observed a moderately strong negative linear relationship between the density and alcohol levels of the red wine dataset. Because we observed a small p-value, we can conclude that it is extremely unlikely for this relationship to have occurred by random chance. Therefore we can reject the null hypothesis.
