```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

red = pd.read_csv('winequality-red.csv', sep=';')
white = pd.read_csv('winequality-white.csv', sep=';')

print(white.columns)
```

```python
#Hypothesis: Red wines will be more acidic than white wines
# Red wines are slightly more acidic that white wines with a mean of 8.31. When compared graphically,
#the difference is not massive and most wines are clustered in the low acity range.
# Red wine does have on average more acidic wines and more very acidic outliers. I can say that red wine is more acidic
#but not to a significant extent.
```

```python
#plt.scatter(red['volatile acidity'], red['fixed acidity'],color='red')
plt.scatter(white['citric acid'], white['fixed acidity'],color='pink', s=1)
plt.scatter(red['citric acid'], red['fixed acidity'],color='red', s=1)
plt.xlabel('citric acid')
plt.ylabel('fixed acidity')
plt.title("citric acid vs fixed acitiy in red and white wines")

print(np.mean(red["citric acid"]))
print(np.mean(white["citric acid"]))

print(np.mean(red["fixed acidity"]))
print(np.mean(white["fixed acidity"]))
```

```python
#Hypothsesis: The quality in red wines will not be affected by pH
#2 histograms, for quality vs pH
np.mean(red['pH']) #=3.3 so assuming 3 is the average pH
high_pH = red['pH'] > 3
low_pH = red['pH'] <= 3.3

plt.subplot(1, 2, 1)
plt.hist(red[low_pH]['quality'], bins=6, color='blue', alpha=0.7)
plt.title('low_pH quality')
plt.xlabel('3 to 8 Quality')

plt.subplot(1, 2, 2)
plt.hist(red[high_pH]['quality'], bins=6, color='blue', alpha=0.7)
plt.title('high_pH quality')
plt.xlabel('3 to 8 Quality')

plt.tight_layout()
```

```python
print(np.mean(red[high_pH]['quality']))
print(np.mean(red[low_pH]['quality']))
```

```python
#Looking at both the histograms, they seem almost identically centered with a bell shaped curve. This means that 
#havign low or high pH doesnt actually affect the quality. The mean of both data sets are also very similar
#so I can furtehr conclude that pH and qualitt do not have any correlation so my hypothesis is wrong
```
